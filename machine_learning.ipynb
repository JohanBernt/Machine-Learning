```python
# convert to dataframe
encoded_df=pd.get_dummies(bank, drop_first='True')
encoded_df.head()
```

```python
# separation of target variable and input variable
Y_data=encoded_df['y_yes']
X_data=encoded_df.drop('y_yes', axis='columns')
```

```python
X_data.shape #get the dimension of the data
X_data.head() # show the head
```

```python
Y_data.shape
Y_data.head()
```

```python
# import required package to test and train model
from sklearn.model_selection import train_test_split
# import package to fit model
from sklearn.neighbors import KNeighborsClassifier
# import confusion matrix package
from sklearn.metrics import confusion_matrix
#import classification report package
from sklearn.metrics import classification_report
#import f1_score package
from sklearn.metrics import f1_score
```

```python
# 50% split KNN=3
x_train, x_test, y_train, y_test = train_test_split(X_data,
                                                    Y_data,
                                                    test_size=0.50,
                                                    random_state=0)
# Select a k-Nearest neighbour classifier - select 3
clf = KNeighborsClassifier(3)
#Fit the data
fit = clf.fit(x_train, y_train)
#Predict on unseen data
predicted = fit.predict(x_test)
# create confusion matrix
confusion_matrix(y_test, predicted)
```

```python
#get and print classification report
report_knn = classification_report(y_test, predicted)
print(report_knn)
```

```python
# use another formula to validate the accuracy score in the table above
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predicted) #looks accurate
```

```python
# used another formula to validate the f1 score in table above
f1_knn = f1_score(y_test, predicted, average='weighted')
print(f1_knn) #looks accurate
```
